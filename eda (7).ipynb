# WORLD HAPPENCESS DATA ANALYSIS

### PROBLEM STATEMENT

One important research problem is to examine the key determinants of happiness across countries and to understand how economic, social, and institutional factors contribute to variations in happiness scores. Despite global development, countries show significant differences in happiness levels, and it remains unclear which factors play the most influential role in shaping these outcomes.
Another research problem focuses on the relationship between income and happiness, specifically analyzing whether higher GDP per capita always leads to higher happiness levels. While economic theory often assumes a positive relationship, real-world data suggests that beyond a certain point, increases in income may have diminishing effects on happiness.

STATISTICAL SUMMARY OF INSIGHTS

* Top Happiest Countries

Countries like Finland, Denmark, and Switzerland have the highest Ladder scores, meaning people report higher life satisfaction.

* Strong Role of Income (GDP)

Higher GDP per capita is strongly linked with higher happiness levels. Wealthier countries tend to rank higher.

* Social Support Matters

Countries with strong social support systems (family, community, welfare) show higher happiness scores.

* Health is Important

Healthy life expectancy is positively related to happiness. People in healthier countries feel more satisfied.

* Freedom Increases Happiness

The ability to make life choices (freedom) significantly improves well-being.

* Corruption Reduces Happiness

Higher perceptions of corruption are associated with lower happiness levels.

* Generosity Has Smaller Impact

While generosity contributes to happiness, its effect is smaller compared to GDP, health, and social support.

* Regional Differences

Western European countries dominate the top rankings, while lower-ranked countries often have weaker economic and social conditions.

* Dystopia Benchmark

The “Dystopia + residual” value shows that even after accounting for known factors, some variation in happiness remains unexplained.


# SCATTER PLOT

```python
import matplotlib.pyplot as plt

# select two numerical columns (change these to your actual column names)
x_col = df_num.columns[0]
y_col = df_num.columns[1]

plt.figure(figsize=(8,6))

plt.scatter(df_num[x_col], df_num[y_col])

plt.title(f'Scatter Plot of {x_col} vs {y_col}')
plt.xlabel(x_col)
plt.ylabel(y_col)

plt.show()
```

# HEAT PLOT

```python
import matplotlib.pyplot as plt
import seaborn as sns

df_num = df.select_dtypes(include=['number'])
corr = df_num.corr()

plt.figure(figsize=(8,6))

sns.heatmap(corr, cmap='coolwarm', linewidths=0.5, annot=True)  # removed annot=True

plt.title('Correlation Heatmap')

plt.show()
```

INSIGHTS OF HEAT PLOT

Strong positive correlations (values close to +1) indicate that some numerical features increase together, suggesting potential relationships or redundancy between variables.

Weak or near-zero correlations show that certain variables have little to no linear relationship, meaning they likely behave independently.
